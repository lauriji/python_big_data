```python
# A. Una función que calcule el precio total de la compra

articulos = {
    'Camisa': 20,
    'Pantalón': 30,
    'Calcetines': 5,
    'Zapatos': 50,
    'Gorra': 10,
    'Bufanda': 15,
    'Gafas': 25,
    'Reloj': 35,
    'Corbata': 40,
}

compra = ['Camisa', 'Pantalón', 'Pantalón', 'Gorra', 'Gafas', 'Corbata']

def precio_total(articulos,compra):

    precio=0

    for index in compra:
        precio += articulos[index]
    return precio
    

precio_total(articulos,compra)
```

```python
# B. Una función que calcule el precio total de la compra con un descuento del 10%
def precio_total_10(articulos,compra):

    precio=0

    for index in compra:
        precio += articulos[index]
    return precio*0.9

precio_total_10(articulos,compra)
```

```python
# C. Una función que calcule el precio total de la compra con un descuento del 10% si la compra supera los 100€

def precio_total_mayor_100(articulos,compra):

    if precio_total(articulos,compra)>100:
        return precio_total(articulos,compra)*0.9
    
    else:
        return precio_total
   

precio_total_mayor_100(articulos,compra)
```

```python
# D. Una función que calcule el precio total aplicando el IVA (21%) 
    #Aquí no aplico el descuento

IVA=0.21

def precio_total(articulos,compra):

    precio=0

    for index in compra:
        precio += articulos[index]
    return precio+precio*IVA
    

precio_total(articulos,compra)
```
